```python
import pandas as pd
import numpy as np

# -----------------------------
# Load
# -----------------------------
annex1 = pd.read_csv("annex1.csv")  # item master
annex2 = pd.read_csv("annex2.csv")  # sales
annex3 = pd.read_csv("annex3.csv")  # wholesale price
annex4 = pd.read_csv("annex4.csv")  # loss rate

# -----------------------------
# Normalize columns + types
# -----------------------------
def clean_cols(df):
    df = df.copy()
    df.columns = df.columns.astype(str).str.strip().str.replace("\u00a0", " ", regex=False)
    return df

annex1 = clean_cols(annex1)
annex2 = clean_cols(annex2)
annex3 = clean_cols(annex3)
annex4 = clean_cols(annex4)

# Item Code as string (safe join, keeps leading zeros)
for df in [annex1, annex2, annex3, annex4]:
    df["Item Code"] = df["Item Code"].astype(str).str.strip()

# Dates
annex2["Date"] = pd.to_datetime(annex2["Date"], errors="coerce")
annex3["Date"] = pd.to_datetime(annex3["Date"], errors="coerce")

# Optional: parse time too (if you want hour/day-part insights)
# If Time is like "07:30:00" or "7:30"
annex2["Time"] = annex2["Time"].astype(str).str.strip()
annex2["sale_datetime"] = pd.to_datetime(
    annex2["Date"].dt.strftime("%Y-%m-%d") + " " + annex2["Time"],
    errors="coerce"
)

# Numeric
annex2["Quantity Sold (kilo)"] = pd.to_numeric(annex2["Quantity Sold (kilo)"], errors="coerce")
annex2["Unit Selling Price (RMB/kg)"] = pd.to_numeric(annex2["Unit Selling Price (RMB/kg)"], errors="coerce")
annex3["Wholesale Price (RMB/kg)"] = pd.to_numeric(annex3["Wholesale Price (RMB/kg)"], errors="coerce")
annex4["Loss Rate (%)"] = pd.to_numeric(annex4["Loss Rate (%)"], errors="coerce")

# Basic validity
annex2.loc[annex2["Quantity Sold (kilo)"] < 0, "Quantity Sold (kilo)"] = np.nan
annex2.loc[annex2["Unit Selling Price (RMB/kg)"] <= 0, "Unit Selling Price (RMB/kg)"] = np.nan
annex3.loc[annex3["Wholesale Price (RMB/kg)"] <= 0, "Wholesale Price (RMB/kg)"] = np.nan
annex4.loc[(annex4["Loss Rate (%)"] < 0) | (annex4["Loss Rate (%)"] > 100), "Loss Rate (%)"] = np.nan

# Dedup masters
annex1 = annex1.drop_duplicates(subset=["Item Code"], keep="first")
annex4 = annex4.drop_duplicates(subset=["Item Code"], keep="first")
annex3 = annex3.drop_duplicates(subset=["Date", "Item Code"], keep="last")

# -----------------------------
# IMPORTANT: Handle "Sale or Return"
# -----------------------------
# If "Return" exists, treat qty as negative (so net sales reflect returns)
annex2["Sale or Return"] = annex2["Sale or Return"].astype(str).str.strip().str.lower()
annex2["signed_qty_kg"] = np.where(
    annex2["Sale or Return"].str.contains("return", na=False),
    -annex2["Quantity Sold (kilo)"],
    annex2["Quantity Sold (kilo)"]
)

# Discount flag
annex2["Discount (Yes/No)"] = annex2["Discount (Yes/No)"].astype(str).str.strip().str.lower()
annex2["discount_flag"] = annex2["Discount (Yes/No)"].isin(["yes", "y", "1", "true"])

# -----------------------------
# MERGE (core answer)
# -----------------------------
merged = (
    annex2
    .merge(
        annex1[["Item Code", "Item Name", "Category Code", "Category Name"]],
        on="Item Code",
        how="left"
    )
    .merge(
        annex3[["Date", "Item Code", "Wholesale Price (RMB/kg)"]],
        on=["Date", "Item Code"],
        how="left"
    )
    .merge(
        annex4[["Item Code", "Loss Rate (%)"]],
        on="Item Code",
        how="left"
    )
)

# -----------------------------
# Derived KPIs for insights
# -----------------------------
merged["loss_rate"] = merged["Loss Rate (%)"] / 100.0

# Revenue (net of returns because signed_qty_kg includes negative)
merged["revenue_rmb"] = merged["signed_qty_kg"] * merged["Unit Selling Price (RMB/kg)"]

# COGS using wholesale price (if missing wholesale, will be NaN)
merged["cogs_rmb"] = merged["signed_qty_kg"] * merged["Wholesale Price (RMB/kg)"]

# Gross margin (simple)
merged["gross_margin_rmb"] = merged["revenue_rmb"] - merged["cogs_rmb"]

# Loss-adjusted cost: if loss_rate=0.05 => need to buy more to sell 1kg
merged["loss_adjusted_cogs_rmb"] = merged["signed_qty_kg"] * (
    merged["Wholesale Price (RMB/kg)"] / (1 - merged["loss_rate"])
)

merged["loss_adjusted_margin_rmb"] = merged["revenue_rmb"] - merged["loss_adjusted_cogs_rmb"]

# -----------------------------
# Insight tables (least sold / remove items / category actions)
# -----------------------------
# 1) Least sold items (by net quantity)
least_sold_items = (
    merged.groupby(["Item Code", "Item Name", "Category Name"], dropna=False)
    .agg(
        net_qty_kg=("signed_qty_kg", "sum"),
        net_revenue=("revenue_rmb", "sum"),
        avg_sell_price=("Unit Selling Price (RMB/kg)", "mean"),
        avg_wholesale=("Wholesale Price (RMB/kg)", "mean"),
        loss_rate=("loss_rate", "mean"),
        loss_adj_margin=("loss_adjusted_margin_rmb", "sum"),
        discount_share=("discount_flag", "mean"),
        txn_count=("Item Code", "size"),
    )
    .reset_index()
    .sort_values(["net_qty_kg", "net_revenue"], ascending=[True, True])
)

# 2) Category performance (which category to push / remove)
category_summary = (
    merged.groupby(["Category Code", "Category Name"], dropna=False)
    .agg(
        net_qty_kg=("signed_qty_kg", "sum"),
        net_revenue=("revenue_rmb", "sum"),
        loss_adj_margin=("loss_adjusted_margin_rmb", "sum"),
        avg_loss_rate=("loss_rate", "mean"),
        discount_share=("discount_flag", "mean"),
        unique_items=("Item Code", "nunique"),
    )
    .reset_index()
    .sort_values("loss_adj_margin", ascending=True)  # low margin categories first
)

# 3) “Remove / Fix” candidates:
# low sales + high loss + negative loss-adjusted margin
remove_candidates = least_sold_items[
    (least_sold_items["net_qty_kg"] <= least_sold_items["net_qty_kg"].quantile(0.25)) &
    (least_sold_items["loss_rate"] >= least_sold_items["loss_rate"].quantile(0.75)) &
    (least_sold_items["loss_adj_margin"] < 0)
].copy()

# -----------------------------
# Save outputs
# -----------------------------
merged.to_csv("merged_sales_price_loss.csv", index=False)
least_sold_items.to_csv("least_sold_items.csv", index=False)
category_summary.to_csv("category_summary.csv", index=False)
remove_candidates.to_csv("remove_candidates.csv", index=False)

print("✅ Saved:")
print(" - merged_sales_price_loss.csv")
print(" - least_sold_items.csv")
print(" - category_summary.csv")
print(" - remove_candidates.csv")
```
